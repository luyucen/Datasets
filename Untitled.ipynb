```python
import numpy as np
import os
import PIL
import PIL.Image as Image
import pathlib
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
import sys

# In Weizmann dataset, there are totally 100 images. 
# Image width is fixed at 300 but image height varies from 125 to 465.
# Here we will resize all images to a fixed size 
img_width = 64
img_height = 64
batch_size = 8
buffer_size = 100
data_dir = pathlib.Path('/home/cly/python_projects/image_segmentation/1obj/split')
```

```python
def input_generator(myset):
  # generate filepaths from any of the three datasets: train, val and test
  # leave placeholders for random images
  if myset == 0:
    image_paths = pathlib.Path(data_dir).glob('train/image/image*.png')
    mask_dir = str(data_dir/'train'/'mask')
  elif myset == 1:
    image_paths = pathlib.Path(data_dir).glob('val/image/image*.png')
    mask_dir = str(data_dir/'val'/'mask')
  else:
    image_paths = pathlib.Path(data_dir).glob('test/image/image*.png')
    mask_dir = str(data_dir/'test'/'mask')
  generated = False
  for image_path in image_paths:
    mask_path = mask_dir + '/mask_' + '_'.join(image_path.name.split('_')[1:])
    if not generated and (np.random.rand() < 0.2):
      generated = True
      yield '', '', generated
    else:
      generated = False
      yield str(image_path), str(mask_path), generated
```

```python
train_ds = tf.data.Dataset.from_generator(input_generator, (tf.string, tf.string, tf.bool), args=[0])
val_ds = tf.data.Dataset.from_generator(input_generator, (tf.string, tf.string, tf.bool), args=[1])
```

```python

# print(tf.data.experimental.cardinality(train_ds).numpy())
# print(tf.data.experimental.cardinality(val_ds).numpy())

def initializeState(img_height, img_width):
    
  center1 = int(img_height / 2)
  center2 = int(img_width / 2)
  r = min(img_height, img_width) / 3
  r2 = r ** 2
  a = np.zeros((img_height, img_width, 1))
  for i in range(img_height):
    for j in range(img_width):
      a[i, j, 0] = float((i - center1) ** 2 + (j - center2) ** 2 < r2)
    
  return a  


def initializePhi(img_height, img_width):
    
  center1 = int(img_height / 2)
  center2 = int(img_width / 2)
  r = min(img_height, img_width)/3
  r2 = r ** 2
  a = np.zeros((img_height, img_width, 1))
  for i in range(img_height):
    for j in range(img_width):
      a[i, j, 0] = float(np.e ** ((-(i - center1) ** 2 + (j - center2) ** 2)/ ((img_height/8) ** 2)))  - 0.5
  return a  

def polar(img_height, img_width):
    # polar coordinate r and theta with respect to the origin at the center of the image
    centerx = int(img_height / 2)
    centery = int(img_width / 2)
    ii = np.arange(img_height)
    jj = np.arange(img_width)
    xx = ii - centerx
    yy = jj - centery
    
    xx = xx.reshape((img_height, 1)) @ np.ones((1, img_width))
    yy = np.ones((img_height, 1)) @ yy.reshape((1, img_width)) 
    
    rr = np.sqrt(np.square(xx) + np.square(yy))
    
    theta = np.zeros((img_height, img_width))
    
    theta[0:centerx, :] = np.arctan(yy[0:centerx, :] / xx[0:centerx]) - np.pi
    theta[centerx, 0:centery] = -np.pi / 2
    theta[centerx, centery+1:] = np.pi / 2
    theta[centerx+1:, :] = np.arctan(yy[centerx+1:, :] / xx[centerx+1:, :])
    
    return rr.reshape(img_height, img_width, 1), theta.reshape(img_height, img_width, 1)

rr, theta = polar(img_height, img_width)

def rose_curve_image(alpha, beta, k):
  return tf.cast(rr < alpha + beta * tf.math.sin(k * theta), dtype = tf.float32) 

def decode_img(img):
  # convert the compressed string to a 3D uint8 tensor of shape [height, width, channels]
  img = tf.image.decode_png(img, channels=1)
  # resize the image to the desired size
  return tf.image.resize(img, [img_height, img_width], method='nearest')

def decode_seg(seg):
  seg = tf.image.decode_png(seg, channels=3)
  # convert the image to a mask
  mask = tf.cast(seg[:, :, 0:1] == tf.reduce_max(seg[:, :, 0]), dtype=tf.float32) 
  
  return tf.image.resize(mask, [img_height, img_width], method='nearest')

def process_path(*argv):
    
  if len(argv) == 1:
    paths = argv[0]
    img_path = paths[0]
    seg_path = paths[1]
    if tf.random.uniform(()) > 0.66:
        seg_path = paths[2]
    elif tf.random.uniform(()) > 0.5:
        seg_path = paths[3]
  else:
    img_path = argv[0]
    seg_path = argv[1]
    
  if len(argv) > 2 and argv[2]:
    alpha = tf.random.uniform(shape=[], maxval=min(img_width, img_height)/2)
    beta = tf.random.uniform(shape=[], maxval=min(img_width, img_height)/2)
    k = tf.cast(tf.random.uniform(shape=[], maxval=7, dtype=tf.int32), dtype=tf.float32) 
    image = rose_curve_image(alpha, beta, k)
    return image, image
  else:
    # load the raw data from the file as a string
    img_string = tf.io.read_file(img_path)
    seg_string = tf.io.read_file(seg_path)
    img = decode_img(img_string)
    mask = decode_seg(seg_string)
    img = tf.image.convert_image_dtype(img, tf.float32)
    return img, mask

def augment(img, mask):
  # Cast and normalize the image to [0,1]
  img = tf.image.random_brightness(img, max_delta=0.5) # Random brightness
  img = tf.image.random_contrast(img, 0.2, 0.5) # Random contrast
  
  # Filp the image and the mask together
  # It seems there is no random rotation in tensorflow?
  # if tf.random.uniform(()) > 0.5:
  #   img = tf.image.flip_up_down(img)
  #   mask = tf.image.flip_up_down(mask)
  # if tf.random.uniform(()) > 0.5:
  #   img = tf.image.flip_left_right(img)
  #   mask = tf.image.flip_left_right(mask)
  return img, mask

AUTOTUNE = tf.data.experimental.AUTOTUNE

train_ds = train_ds.map(process_path, num_parallel_calls=AUTOTUNE)
val_ds = val_ds.map(process_path, num_parallel_calls=AUTOTUNE)

for image, mask in train_ds.take(1):
  print("Image shape: ", image.numpy().shape)
  print("Mask shape: ", mask.numpy().shape)
  print("Image max:", np.max(image.numpy()))
  print("Image min:", np.min(image.numpy()))

def configure_for_performance(ds):
  ds = ds.shuffle(buffer_size=buffer_size)
  ds = ds.batch(batch_size)
  ds = ds.prefetch(buffer_size=AUTOTUNE)
  return ds

train_ds = train_ds.cache()
val_ds = val_ds.cache()

train_ds = train_ds.map(augment, num_parallel_calls=AUTOTUNE)
train_ds = configure_for_performance(train_ds)
val_ds = configure_for_performance(val_ds)

```

```python
class fidelityTerm(keras.layers.Layer):
  def __init__(self, img_size, threshold=0.5):
    super(fidelityTerm, self).__init__()
    self.Ug = self.add_weight(shape=(img_size, img_size), initializer=tf.random_normal_initializer(mean=1), trainable=True)
    self.Wg = self.add_weight(shape=(img_size, img_size), initializer=tf.random_normal_initializer(mean=1), trainable=True)
    self.flatten = keras.layers.Flatten()
    self.threshold = threshold
    
  def call(self, inputs):
    img = inputs[0]
    state = inputs[1]
    
    # c1 = tf.stop_gradient(tf.divide(tf.reduce_sum(img*state, axis=1), tf.reduce_sum(state, axis=1)))
    # c2 = tf.stop_gradient(tf.divide(tf.reduce_sum(img*(1-state), axis=1), tf.reduce_sum(1-state, axis=1)))
    c1 = tf.stop_gradient(tf.divide(tf.reduce_sum(tf.cast(state > self.threshold, tf.float32) * img, axis=1), tf.reduce_sum(tf.cast(state > 0.5, tf.float32), axis=1)+sys.float_info.epsilon))
    c2 = tf.stop_gradient(tf.divide(tf.reduce_sum(tf.cast(state < self.threshold, tf.float32) * img, axis=1), tf.reduce_sum(tf.cast(state < 0.5, tf.float32), axis=1)+sys.float_info.epsilon))
    
    # reshape c1 from [batch_size, ] to [batch_size, 1]
    c1 = self.flatten(c1)
    c2 = self.flatten(c2)
    
    return tf.matmul(tf.math.squared_difference(img, c1), self.Ug) - tf.matmul(tf.math.squared_difference(img, c2), self.Wg)

class myModelBlock(keras.layers.Layer):
  def __init__(self, img_height, img_width):
    super(myModelBlock, self).__init__()
    self.fidelity = fidelityTerm(img_height*img_width)
    self.reshape2D = keras.layers.Reshape((img_height, img_width, 1))
    self.conv2D = keras.layers.Conv2D(1, 3, padding='same')
    self.flatten = keras.layers.Flatten()
    self.dense = keras.layers.Dense(img_height*img_width)
    self.activation = keras.layers.Activation(activation=tf.keras.activations.tanh)
  
  def call(self, inputs):
    state = inputs[1]
    x = self.fidelity(inputs)
    y = self.reshape2D(state)
    y = self.conv2D(1 - 2 * y)
    y = self.flatten(y)
    y = self.dense(- x - y)
    return [inputs[0], 0.5 + 0.5 * self.activation(y)]

class myModel(keras.Model):
  def __init__(self, img_height, img_width, state0):
    super(myModel, self).__init__()
    self.state0 = state0
    self.flatten = keras.layers.Flatten()
    self.block1 = myModelBlock(img_height, img_width)
    self.block2 = myModelBlock(img_height, img_width)
    self.block3 = myModelBlock(img_height, img_width)
    self.reshape = keras.layers.Reshape((img_height, img_width, 1))
    
  def call(self, inputs):
    x = self.flatten(inputs)
    x = self.block1([x, tf.reshape(state0, [1, -1])])
    x = self.block2(x)
    x = self.block3(x)[1]
    return self.reshape(x)
    
    
```

```python
class fidelityTermHeaviside(keras.layers.Layer):
  def __init__(self, img_height, img_width, eps=1e-10):
    super(fidelityTermHeaviside, self).__init__()
    self.eps = eps
    img_size = img_height*img_width
    self.img_height = img_height
    self.img_width = img_width
    self.Ug = self.add_weight(shape=(img_size, img_size), initializer=tf.random_uniform_initializer(-1/img_height, 1/img_height), trainable=True)
    self.Wg = self.add_weight(shape=(img_size, img_size), initializer=tf.random_uniform_initializer(-1/img_height, 1/img_height), trainable=True)
    self.bg = self.add_weight(shape=(img_size,), initializer=tf.random_uniform_initializer(-1/img_size, 1/img_size), trainable=True)
    self.reshape2D = keras.layers.Reshape((img_height, img_width, 1))
    self.flatten = keras.layers.Flatten()
    self.xgradop, self.ygradop = self.gradientops()
       
  def call(self, inputs):
    
    img = inputs[0]
    state = inputs[1]
    
    c1, c2 = self.means(img, state)
    kappa = self.kappa(state)
    fidelity = -tf.matmul(tf.math.squared_difference(img, c1), self.Ug) + tf.matmul(tf.math.squared_difference(img, c2), self.Wg) + self.bg
    
    return kappa + fidelity/(tf.reduce_max(fidelity) + sys.float_info.epsilon)
    
    
  def means(self, img, state):
    
    in_idx = tf.cast(state >= 0, tf.float32)
    out_idx = tf.cast(state < 0, tf.float32)
    
    # Piecewise Heaviside
    H = 1 / 2 * (1 + state / self.eps + (1 / np.pi) * tf.math.sin(state * np.pi / self.eps))
    H_n = 1 - H
    H1 = tf.cast(state > self.eps, tf.float32)
    H3_1 = tf.cast(state < self.eps, tf.float32) * tf.cast(state > -self.eps, tf.float32)
    H2 = 1 - H1
    H3 = H * H3_1
    H4 = H_n * H3_1
    
    # for the inside
    sum1 = tf.reduce_sum(img * H1, axis=1)
    sum2 = tf.reduce_sum(img * H3, axis=1)
    c1 = tf.stop_gradient((sum1 + sum2) / (tf.reduce_sum(
            in_idx, axis = 1) + sys.float_info.epsilon) )  # (sum1 + sum2)/(np.sum(idx1) + np.sum(idx3) + self.eps)

    # for the outside ?
    sum1 = tf.reduce_sum(img * H2, axis=1)
    sum2 = tf.reduce_sum(img * H4, axis=1)
    c2 = tf.stop_gradient((sum1 + sum2) / (tf.reduce_sum(
            out_idx, axis = 1) + sys.float_info.epsilon)) # (sum1 + sum2)/(np.sum(idx2) + np.sum(idx3) + self.eps)
    
    return self.flatten(c1), self.flatten(c2)


  def kappa(self, state):  
        
    statex, statey = self.gradient(self.reshape2D(state))
    statexx, statexy = self.gradient(statex)
    _, stateyy = self.gradient(statey)
    
    kappa = tf.stop_gradient(statexx * statey * statey - 2 * statex * statey * statexy + stateyy * statex * statex)
    kappa = tf.stop_gradient(kappa/(tf.pow(statex * statex + statey * statey, 1.5) + sys.float_info.epsilon))
    
    # normalize kappa
    kappa = kappa/(tf.reduce_max(kappa) + sys.float_info.epsilon)
    
    return self.flatten(kappa)
  
    
  def gradientops(self):
  
    ygradop = np.zeros((self.img_height, self.img_width))
    
    for i in range(1, img_height - 1):
      ygradop[i, i-1:i+2] = [-0.5, 0, 0.5]
    
    ygradop[0, 0:2] = [-1, 1]
    ygradop[img_height-1, img_width-2:img_width] = [-1, 1]
    
    if img_height == img_width:
        
      xgradop = np.transpose(ygradop)
    
    else:
        
      xgradop = np.zeros(img_width, img_height)
        
      for j in range(1, img_width - 1):
        xgradop[j-1:j+2, j] = [-0.5, 0, 0.5]
        
      xgradop[0:2, 0] = [-1, 1]
      xgradop[img_height-2:img_height, img_width - 1] = [-1, 1]

    return xgradop, ygradop

  def gradient(self, x):
    
    dx = tf.einsum('aikp,kj->aijp', x, self.xgradop)
    dy = tf.einsum('ik,akjp->aijp', self.ygradop, x)

    return dx, dy
        

class GRU(keras.layers.Layer):
  def __init__(self, img_height, img_width):
    
    super(GRU, self).__init__()
    
    img_size = img_height * img_width
    
    self.Uz = self.add_weight(shape=(img_size, img_size), initializer='random_normal', trainable=True, name='Uz')
    self.Wz = self.add_weight(shape=(img_size, img_size), initializer='random_normal', trainable=True, name='Wz')
    self.bz = self.add_weight(shape=(img_size,), initializer=tf.random_uniform_initializer(-1/img_size, 1/img_size), trainable=True, name='bz')
    self.Ur = self.add_weight(shape=(img_size, img_size), initializer='random_normal', trainable=True, name='Ur')
    self.Wr = self.add_weight(shape=(img_size, img_size), initializer='random_normal', trainable=True, name='Wr')
    self.br = self.add_weight(shape=(img_size,), initializer=tf.random_uniform_initializer(-1/img_size, 1/img_size), trainable=True, name='br')
    self.Uy = self.add_weight(shape=(img_size, img_size), initializer='random_normal', trainable=True, name='Uy')
    self.Wy = self.add_weight(shape=(img_size, img_size), initializer='random_normal', trainable=True, name='Wy')
    self.by = self.add_weight(shape=(img_size,), initializer=tf.random_uniform_initializer(-1/img_size, 1/img_size), trainable=True, name='by')
    
    self.fidelity = fidelityTermHeaviside(img_height, img_width)
    self.sigmoid = keras.layers.Activation(activation=tf.nn.sigmoid)
    self.tanh = keras.layers.Activation(activation=tf.keras.activations.tanh)

  def call(self, inputs):
    
    x = self.fidelity(inputs)
    # x = inputs[0]
    state = inputs[1]
    
    zt = self.sigmoid(tf.matmul(x, self.Uz) + tf.matmul(state, self.Wz) + self.bz)
    rt = self.sigmoid(tf.matmul(x, self.Ur) + tf.matmul(state, self.Wr) + self.br)
    ht = self.tanh(tf.matmul(x, self.Uy) + tf.matmul(state * rt, self.Wy) + self.by)
    
    return [inputs[0], zt * state + (1 - zt) * ht]

class levelsetModel(keras.Model):
  def __init__(self, img_height, img_width, state0):
    
    super(levelsetModel, self).__init__()
    
    self.state0 = state0
    
    self.GRU1 = GRU(img_height, img_width)
    self.GRU2 = GRU(img_height, img_width)
    self.GRU3 = GRU(img_height, img_width)
    
    self.flatten = keras.layers.Flatten()
    self.reshape = keras.layers.Reshape((img_height, img_width, 1))
    self.dense = keras.layers.Dense(img_height * img_width, activation=tf.nn.sigmoid)
                             
  def call(self, inputs):
    x = self.flatten(inputs)
    x = self.GRU1([x, tf.reshape(state0, [1, -1])])
    x = self.GRU2(x)
    x = self.GRU3(x) 
    x = self.dense(x[1])
    return self.reshape(x)
    
```

```python
state0 = initializeState(img_height, img_width)
model = myModel(img_height, img_width, state0)
optimizer = keras.optimizers.Adam(learning_rate=1e-3)
model.compile(optimizer, loss=keras.losses.BinaryCrossentropy())
model.fit(train_ds, validation_data=val_ds, epochs=5)
```
